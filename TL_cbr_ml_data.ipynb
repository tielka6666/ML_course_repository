## Sklearn.datasets <a name='datasets'></a>

документация: http://scikit-learn.org/stable/datasets/

```python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.pyplot.style.use('ggplot')

import pandas as pd
from sklearn import datasets
import seaborn as sns
```

```python
x, y = datasets.make_classification(n_samples=100, n_features=4, n_informative=2, n_redundant=0, n_classes=3, 
                                    n_clusters_per_class=1)
```

```python
plt.figure(figsize=(8, 8))
plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)

plt.subplot(331)
plt.title("features 0 and 1", fontsize='small')
plt.scatter(x[:,0], x[:,1], c=y, cmap = matplotlib.colors.ListedColormap(['red','green','black']))

plt.subplot(332)
plt.title("features 0 and 2", fontsize='small')
plt.scatter(x[:,0], x[:,2], c=y, cmap = matplotlib.colors.ListedColormap(['red','green','black']))

plt.subplot(333)
plt.title("features 0 and 3", fontsize='small')
plt.scatter(x[:,0], x[:,3], c=y, cmap = matplotlib.colors.ListedColormap(['red','green','black']))
```
